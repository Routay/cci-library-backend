# Jour 9 - 23 Juillet 2026 : Visualisation avec Matplotlib

Un graphique vaut mille nombres ! **Matplotlib** est la bibliothèque de base pour dessiner des courbes et des graphiques en Python.

## 1. Courbe simple (Line plot)
Idéal pour montrer l'évolution d'une donnée dans le temps.

```python
import matplotlib.pyplot as plt

annees = [2018, 2019, 2020, 2021, 2022, 2023]
ventes = [150, 180, 160, 210, 250, 280]

plt.plot(annees, ventes, color='blue', marker='o')
plt.title("Évolution des ventes")
plt.xlabel("Années")
plt.ylabel("Nombre de ventes")
plt.show() # Affiche le graphique
```

## 2. Diagramme en barres
Parfait pour comparer des catégories distinctes.

```python
produits = ['Pommes', 'Bananes', 'Cerises']
quantites = [45, 30, 15]

plt.bar(produits, quantites, color=['red', 'yellow', 'pink'])
plt.title("Stock de fruits")
plt.show()
```
